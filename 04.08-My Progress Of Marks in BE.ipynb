# Visualaizing My Progress Of Marks in BE | Predicting the Marks using LinearRegression

```python
import matplotlib.pyplot as plt  
%matplotlib inline

def avg(a,b):
    return (a+b)/2

j = [avg(7.04,7.00),avg(6.89,7.25),avg(6.63,7.69)]
k = ['2nd Year','3rd Year','4th Year']

plt.figure(figsize=(10,5))
plt.plot(k,j,'-s',linewidth=2,markersize=12,markerfacecolor='white',markeredgewidth=3,markeredgecolor='green')
plt.ylabel('CGPA',fontsize=15)
plt.xlabel('Years',fontsize=15)
plt.title('Progress Of Marks in BE',fontsize=25)

#plt.rcParams['axes.spines.left'] = False
plt.rcParams['axes.spines.right'] = False
plt.rcParams['axes.spines.top'] = False
#plt.rcParams['axes.spines.bottom'] = False
```

```python
import pandas as pd, numpy as np

df = pd.concat([pd.DataFrame(np.array([2,3,4]),columns=['No of Years']),pd.DataFrame(np.array(j),columns=['CGPA'])],axis=1)
df
```

# LinearRegression

```python
# LinearRegression

# Splitting Indepedent & Dependent variables
X = df[['No of Years']]
y = df[['CGPA']]

# Since i have to train all the data points so
X_train = X
y_train = y

# Fitting linearRegression model
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train,y_train)
```

# Visualaizing the train set results

```python
plt.figure(figsize=(10,5))

plt.scatter(X_train,y_train,color='red')
plt.plot(X_train,lm.predict(X_train),color='blue')

plt.ylabel('CGPA',fontsize=15)
plt.xlabel('Years',fontsize=15)
plt.title('Progress Of Marks in BE',fontsize=25)

#plt.rcParams['axes.spines.left'] = False
plt.rcParams['axes.spines.right'] = False
plt.rcParams['axes.spines.top'] = False
#plt.rcParams['axes.spines.bottom'] = False
```

# Lets Predict...!

```python
X_test = np.array(5).reshape(-1,1)

prediction = lm.predict(X_test)

prediction
```
