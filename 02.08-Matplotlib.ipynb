## Importing required libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=['java','python','php','javascript','c#','C++']
y=[22.2,17.6,8.8,8,7.7,6.7]

bar=plt.bar(x,y,color=['red','blue','black','green','blue','yellow'])
plt.xlabel('languages',size=14)
plt.ylabel('popularity',size=16)
plt.title('programming languages')

plt.show()
```

```python
index=np.arange(len(x))
index
```

```python
bar=plt.bar(index,y,color=['red','blue','black','green','blue','yellow'])

plt.xticks(index,x)       # inplace of x we can put anything or we can pass a direct new list also...

plt.xlabel('languages',size=14)
plt.ylabel('popularity',size=16)
plt.title('programming languages')

plt.show()
```

```python
a=np.array([0,1,2,3])
b=np.array([100,200,300,400])
c=np.array([300,355,655,855])

plt.plot(a,b,linestyle='--',linewidth=4,color='blue',label='cos func')
plt.plot(a,c,linestyle=':',linewidth=4,color='green',label='exp*cos func')
plt.xlabel('me')
plt.ylabel('you')
plt.legend()

plt.show()
```

```python
# Sample Histogram(it takes only one argument)

x=np.arange(0,10,0.1)
y1=(((x-3)**3)-100)+np.random.randint(-20,20,size=len(x))

plt.hist(y1)

plt.show
```

```python
# Scatter plots

plt.scatter(a,b,color='red')
plt.scatter(a,c,color='blue')

plt.show()
```

```python
# Subplots

fig, axes=plt.subplots(nrows=2,ncols=2)
```
